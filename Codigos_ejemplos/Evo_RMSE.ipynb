# Evolución Modelo de Predicción según lag aplicado

```python
import numpy as np
import pandas as pd
```

```python
lag3 = pd.read_csv('/content/RMSE_demanda_lag3.csv')
lag4 = pd.read_csv('/content/RMSE_demanda_lag4.csv')
lag5 = pd.read_csv('/content/RMSE_demanda_lag5.csv')
lag6 = pd.read_csv('/content/RMSE_demanda_lag6.csv')
lag9 = pd.read_csv('/content/RMSE_demanda_lag9.csv')
lag12 = pd.read_csv('/content/RMSE_demanda_lag12.csv')
```

```python
# Realizas el merge de lag3 y lag4 en base a la columna 'RubroN3'
merged_df = pd.merge(lag3, lag4, on='RubroN3', how='outer')
merged_df = pd.merge(merged_df, lag5, on='RubroN3', how='outer')
merged_df = pd.merge(merged_df, lag6, on='RubroN3', how='outer')
merged_df = pd.merge(merged_df, lag9, on='RubroN3', how='outer')
merged_df = pd.merge(merged_df, lag12, on='RubroN3', how='outer')
data = merged_df.copy()
data = data.replace([np.inf, -np.inf], np.nan).dropna()

data['mean'] = data[['lag3', 'lag4', 'lag5', 'lag6', 'lag9', 'lag12']].mean(axis=1)
datalag3 = data.sort_values(by='lag3').head(20)
datalag3
```

```python
visualizacion = data[['RubroN3','lag3']].sort_values(by='lag3')

# Renombrar la columna 'lag3' a 'RMSE relativo'
visualizacion.rename(columns={'lag3': 'RMSE relativo'}, inplace=True)

# Convertir los valores de la columna a formato porcentual con dos decimales
visualizacion['RMSE relativo'] = (visualizacion['RMSE relativo'] * 100).round(2).astype(str) + '%'

# Mostrar la tabla resultante
visualizacion
```

```python
import pandas as pd
from tabulate import tabulate

# Crear la tabla en formato LaTeX
latex_table = tabulate(visualizacion, tablefmt='latex', headers='keys', showindex=False)

# Agregar el título de la tabla
table_title = 'Tabla de RMSE Relativo por Clase'
latex_table_with_title = r'\begin{table}' + '\n' + \
                        r'\caption{' + table_title + '}' + '\n' + \
                        r'\resizebox{0.9\textwidth}{!}{' + '\n' + latex_table + '\n' + r'}' + '\n' + \
                        r'\end{table}'

# Imprimir la tabla en formato LaTeX con el título
print(latex_table_with_title)
```

```python
import plotly.graph_objects as go

# Calcula el promedio de las columnas
promedio_valores = data[['lag3', 'lag4','lag5','lag6','lag9','lag12']].mean()

# Obtén los nombres de las variables
variables = promedio_valores.index

# Configura la figura
fig = go.Figure()

# Agrega las barras al gráfico
for variable, valor in zip(variables, promedio_valores):
    fig.add_trace(go.Bar(
        x=[variable],
        y=[valor],
        name=variable,
        text=[f'{round(valor * 100, 2)}%'],  # Valor en formato de porcentaje
        textposition='auto',  # Coloca el texto automáticamente
        marker=dict(color='steelblue')
    ))

# Configura el diseño del gráfico
fig.update_layout(
    title='Promedio de RMSE para diferentes variables',
    xaxis=dict(
        title='Variables',
        tickangle=45,
        tickfont=dict(size=12),
        showgrid=False,
        zeroline=False,
        linecolor='black',
        linewidth=1
    ),
    yaxis=dict(
        title='Promedio',
        showgrid=True,
        gridcolor='lightgray',
        zeroline=False,
        linecolor='black',
        linewidth=1,
        tickfont=dict(size=12),
        tickformat='.0%'
    ),
    showlegend=True,
    legend=dict(x=0, y=1),
    bargap=0.3,  # Espacio entre las barras
    plot_bgcolor='white',
    margin=dict(l=60, r=20, t=60, b=20)  # Margen alrededor del gráfico
)

# Muestra el gráfico
fig.show()
```

```python
import plotly.express as px

# Ordenar el DataFrame por el valor de 'lag3'
data_sorted = data.sort_values(by='lag3').head(20)

# Truncar los nombres de las clases a un máximo de 40 caracteres
data_sorted['RubroN3'] = data_sorted['RubroN3'].apply(lambda x: x[:30] + '...' if len(x) > 40 else x)

# Crear la gráfica de barras
fig = px.bar(data_sorted, x='RubroN3', y='lag3', labels={'lag3': 'Lag 3 (%)'})

# Configurar el eje Y para mostrar los valores en formato porcentual con un decimal
fig.update_layout(
    title="Valor del Lag 3 para cada Clase",
    xaxis_title="Clase",
    yaxis_title="RMSE relativo",
    xaxis_tickangle=-45,
    yaxis_tickformat='.1%',  # Formato de porcentaje con un decimal
    margin=dict(b=100)  # Ajustar el margen inferior (b) para dejar espacio para el título del eje X
)

# Agregar etiquetas (textos) para mostrar los valores de las barras
fig.update_traces(texttemplate='%{y:.1%}', textposition='outside')

# Mostrar la gráfica
fig.show()
```
